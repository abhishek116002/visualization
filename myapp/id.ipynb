```python
os.chdir(base+'visualization/myapp/id')

names = [
    'co_code','name','owner','industry_type','incorporation_year','nse', 'bse'
        ]

id = pd.read_csv(
    'id_dat.txt',sep='|', header=0,
    names=names,  usecols=[0,1,3,5,6,7,8]
)


## firms whose industry type is NaN, have no data in StandAloneFinancialStatements. Hence, remove those firms.
id = id[id.industry_type.notnull()]

id[id.nse == 'INFY']
```

```python
## remove all firms that are controlled by the govt.

ownership = pd.DataFrame(
    ['Central Govt. - Commercial Enterprises', 
     'State Govt. - Commercial Enterprises',
     'Central Government - Takenover Enterprises', 
     'State Govt. - Statutory Bodies', 'State Government',   
     'Central and Private sector', 'State and Private sector', 
     'State Bank of India Group'
    ], 
    columns = ['owner']   )


id = id[(~( id.owner.isin(ownership.owner) ))]#.drop(labels = ['owner'], axis = 1)
id
```

```python
## sure_id contains data on firms that should be taken for sure.
sure_id = pd.merge(left = id, right = industry_type[industry_type.keep == 1], how = 'inner', on = 'industry_type')
sure_id.drop(columns = ['keep','bse','owner'], inplace = True)
## remove all rows with any NaN values
sure_id.dropna(axis=0, how='any', inplace=True)
sure_id#[sure_id.co_code == 98907]
```

```python
## save the sure_id dataframe into csv file.

os.chdir('/home/abhishek/Documents/visualization/myapp/id')
sure_id.to_csv('sure_id.csv', index = False)
```
